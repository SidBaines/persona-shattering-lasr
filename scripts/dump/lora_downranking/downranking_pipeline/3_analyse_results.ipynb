```python
from pathlib import Path
import json

scores_results = json.loads(Path("./2_judge_scores.json").read_text())
len(scores_results)
```

```python
scores_results
```

```python
def downrank_config_string_to_parts(downrank_config_string):
    persona_config, rest_config = downrank_config_string.split("_rank_")
    rest_config, judge_index = rest_config.split("_judgecall_")
    rest_config, question_index = rest_config.split("_questionindex_")
    persona = persona_config.split("_")[1]
    rank, _, keep_frac = rest_config.split("_")
    return persona, int(rank), float(keep_frac), int(question_index), int(judge_index)
```

```python
import pandas as pd

all_res = []
for key, score_dict in scores_results["scores"].items():
    persona, rank, keep_frac, question_index, judge_index = downrank_config_string_to_parts(key)
    all_res.append({
        "persona": persona,
        "rank": rank,
        "keep_frac": keep_frac,
        "question_index": question_index,
        "judge_index": judge_index,
        "score": score_dict[persona],
    })

res_df = pd.DataFrame(all_res)
res_df
```

```python
res_df["label"] = "rank_" + res_df["rank"].astype(str) + "_keep_frac_" + res_df["keep_frac"].astype(str)
```

```python
import plotly.express as px

for persona, persona_df in res_df.groupby("persona"):
    print(f"Persona: {persona}")

    fig = px.box(
        persona_df,
        x="label",
        y="score",
        color="question_index",
        boxmode="group",
    )

    fig.update_layout(
        title="Score Distribution by Label and Question Index",
        xaxis_title="Label",
        yaxis_title="Score",
        legend_title="Question Index",
    )

    fig.show()
    fig.write_html(f"./{persona}.html")
```
